```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### Line Chart
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)  # plot() : 그래프를 그린다..
# plt.show()
```

```python
x = [1,2,3,4,5]
y = [i**2 for i in x]

x1 = [1,2,3,4,5]
y1 = [6,7,8,9,10]

plt.plot(x1, y1)
plt.plot(x, y)
```

```python
x1 = range(50)
y1 = [np.cos(v) for v in x1]

x2 = range(50)
y2 = [np.sin(v) for v in x2]

plt.plot(x1, y1)
plt.plot(x2, y2)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)  # 도화지를 1행 2열로 나누고 첫번째 
plt.plot(x1, y1, color='red', linestyle='dotted')

plt.subplot(1,2,2) # 도화지를 1행 2열로 나누고 두번째 
plt.plot(x2, y2, color='green', linestyle='dashed')
```

```python
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value+50 for value in x2]

plt.plot(x1, y1, c='#FF6600', ls=':', label='Line 1')
plt.plot(x2, y2, c='#CCCCFF', ls='-', label='Line 2')

plt.text(50, 60, 'Line 1')
plt.annotate('Line 2', xytext = (30,110), xy=(50,100), arrowprops=dict(facecolor='green', shrink=0.1))

plt.legend(fancybox=True, shadow=True, loc='upper right')

plt.xlabel('X_axis')
plt.ylabel('Y_axis')
plt.title('Line Chart')

plt.savefig('matplotlib_line.png')
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

colors = np.random.rand(n)
area = np.pi *  (30*np.random.rand(n))**2

# plt.scatter(x,y)
plt.scatter(x,y, s=area, c=colors, alpha=0.5)

plt.savefig('random_scatter.png')
```
