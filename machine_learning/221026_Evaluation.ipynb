```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.base import BaseEstimator

class MyDummyClassifier(BaseEstimator):
    
    def fit(self, x, y=None):
        pass
    
    def predict(self, x):
        pred = np.zeros((x.shape[0], 1))
        
        for i in range(x.shape[0]):
            if x['Sex'].iloc[i] == 1:
                pred[i] = 0
            else:
                pred[i]  = 1
                
        return pred
```

```python
from sklearn.preprocessing import LabelEncoder

def fillna(df):
    df['Age'].fillna(df['Age'].mean(), inplace=True)
    df['Cabin'].fillna('N', inplace=True)
    df['Embarked'].fillna('N', inplace=True)
    df['Fare'].fillna(0, inplace=True)
    return df

def drop_features(df):
    df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
    return df

def format_features(df):
    df['Cabin'] = df['Cabin'].str[:1]
    features = ['Cabin', 'Sex', 'Embarked']
    for feature in features:
        le = LabelEncoder()
        le = le.fit(df[feature])
        df[feature] = le.transform(df[feature])
    return df

def transform_features(df):
    df = fillna(df)
    df = drop_features(df)
    df = format_features(df)
    return df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

titanic_df = pd.read_csv('./ml_data/titanic_data.csv')
y_titanic_df = titanic_df['Survived']
```

```python
titanic_df.head(2)
```

```python
x_titanic_df = titanic_df.drop('Survived', axis=1)
x_titanic_df = transform_features(x_titanic_df)
x_titanic_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_titanic_df, y_titanic_df, test_size=0.2)
```

```python
myclf = MyDummyClassifier()
myclf.fit(x_train, y_train)

mypredictions = myclf.predict(x_test)
accuracy_score(y_test, mypredictions)
```

```python
from sklearn.datasets import load_digits
from sklearn.base import BaseEstimator
```

```python
class MyFakeClassifier(BaseEstimator):
    def fit(self, x, y):
        pass
    
    def predict(self, x):
        return np.zeros((len(x),1), dtype=bool)
```

```python
digits = load_digits()
```

```python
digits.data
```

```python
digits.data.shape  # 0~9   64 vector ,   1797 ...
```

```python
digits.target  
```

```python
len(digits.target)
```

```python
digits.target == 7
```

```python
y = (digits.target == 7).astype(int)
```

```python
x = digits.data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=11)
```

```python
fakeclf = MyFakeClassifier()
fakeclf.fit(x_train, y_train)
fakepred = fakeclf.predict(x_test)
```

```python
accuracy_score(y_test, fakepred)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, fakepred)
```

```python
print(classification_report(y_test, fakepred))
```
