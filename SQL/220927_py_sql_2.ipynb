```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# gmarket bestseller에서 제목, url, old 가격, discount 가격, 할인율을 출력하시오..
```

```python
url = 'https://corners.gmarket.co.kr/Bestsellers'
```

```python
res = requests.get(url)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
lis = soup.select('div.best-list > ul > li')
```

```python
# 한 페이지(베스트셀러 메인페이지) 내에서 랭킹, 제목, 링크주소, 가격, 할인율

for li in lis:
    print(li.select('p')[0].text)   # 랭킹
    print(li.select('a.itemname')[0].text)   # 타이틀
    print(li.select('a.itemname')[0]['href'])   # 상세 페이지 url
    
    
    print(li.select('div.s-price> strong > span')[0].text)   # 가격
    
    try:
        print(li.select('div.s-price > span >em')[0].text)   # discount
    except:
        print('no sale')  # discount가 없으면 실행
        
    print()
```

```python
# 페이지 url

cate_urls = []
main_url = 'https://corners.gmarket.co.kr'

lis = soup.select('ul#categoryTabG > li')
for li in lis[1:]:
    print(main_url + li.select('a')[0]['href'])
    cate_urls.append(main_url + li.select('a')[0]['href'])
```

```python
for url in cate_urls:
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    lis = soup.select('div.best-list > ul > li')
    
    for li in lis:
        print(li.select('p')[0].text)
        print(li.select('a.itemname')[0].text)
        print(li.select('a.itemname')[0]['href'])
        print(li.select('div.s-price> strong > span')[0].text)

        try:
            print(li.select('div.s-price > span >em')[0].text)
        except:
            print('no sale')
        print()
        
    print('-----------------------------')
    
```

```python
sql = """
drop table if exists best_prd;

create table best_prd(
ranking int,
title varchar(100),
url varchar(150),
price int,
discount int
);
"""

cursor.execute(sql)
```

```python
ranking = 1
title = '[매일유업]매일유업 요미요미 유기농 주스 125ml 24팩 모음'
url = "http://item.gmarket.co.kr/Item?goodscode=2586810618&ver=637998846875306146"
price = int('16,500원'[:-1].replace(',', ''))
discount = int('29%'[:-1])

print(ranking, title, url, price, discount)
```

```python
sql = "insert into best_prd values ({}, '{}', '{}', {}, {});".format(ranking, title, url, price, discount)

cursor.execute(sql)
```

```python
pd.read_sql('select * from best_prd;', db)
```

```python
for url in cate_urls:
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    lis = soup.select('div.best-list > ul > li')
    
    for li in lis:
        ranking = li.select('p')[0].text
        title = li.select('a.itemname')[0].text
        url = li.select('a.itemname')[0]['href']
        price = int(li.select('div.s-price> strong > span')[0].text[:-1].replace(',', ''))

        try:
            discount = int(li.select('div.s-price > span >em')[0].text[:-1])
        except:
            discount = 0
            
        sql = "insert into best_prd values ({}, '{}', '{}', {}, {})\
        ;".format(ranking, title, url, price, discount)
        
        cursor.execute(sql)
```
